# 28. 경제 시계열 예측 프로젝트 포트폴리오 문서 만들기

이번 장에서는 경제 시계열 예측 프로젝트를 포트폴리오 문서로 정리합니다.

핵심 흐름:

```text
API 데이터 수집
-> target/lag 설계
-> 미래 정보 누출 방지
-> MLP/LSTM 비교
-> 한계와 개선 방향
```

## 1. 프로젝트 기본 정보

````python
project_info = {
    "title": "FRED 경제 지표 기반 다음 달 CPI 예측 프로젝트",
    "dataset": "FRED API 경제 지표",
    "target": "target_next_cpi",
    "problem_type": "시계열 회귀",
    "models": ["MLP", "LSTM"],
    "metrics": ["MAE", "RMSE"],
}

project_info
````

## 2. 문제 정의

````python
problem_definition = """FRED API로 수집한 경제 지표를 사용해 다음 달 CPI를 예측하는 시계열 회귀 프로젝트를 구축한다.
CPI, 실업률, 연방기금금리 등 월별 경제 지표를 입력으로 사용하고, lag feature 기반 MLP와 window 기반 LSTM을 비교한다.
이 프로젝트의 목적은 경제 시계열 데이터 수집, 예측 문제 설계, 딥러닝 모델 비교 흐름을 포트폴리오화하는 것이다."""

print(problem_definition)
````

## 3. 데이터 수집 설명

````python
data_collection_description = """FRED API를 사용해 CPI, 실업률, 연방기금금리 데이터를 수집했다.
API 키는 .env 파일로 관리해 코드에 직접 노출하지 않았다.
API 응답 JSON을 pandas DataFrame으로 변환하고, 날짜와 숫자 타입을 정리한 뒤 CSV로 저장했다."""

print(data_collection_description)
````

## 4. 데이터셋 설계 설명

````python
dataset_design_description = """예측 target은 다음 달 CPI이며, pandas shift(-1)을 사용해 target_next_cpi 컬럼을 만들었다.
입력 특성에는 현재 경제 지표와 cpi_lag_1, cpi_lag_2, unemployment_rate_lag_1 같은 과거 lag feature를 포함했다.
이 설계는 현재와 과거 정보를 사용해 다음 달 CPI를 예측하는 구조다."""

print(dataset_design_description)
````

## 5. 미래 정보 누출 방지 설명

````python
leakage_prevention_description = """입력 X에는 현재와 과거 값만 사용하고, 다음 달 CPI는 정답 y로만 사용했다.
train/validation 데이터는 무작위로 섞지 않고 시간 순서대로 나누었다.
스케일러는 train 데이터에만 fit하고 validation 데이터에는 transform만 적용해 미래 정보 누출을 줄였다."""

print(leakage_prevention_description)
````

## 6. 모델 비교 설명

````python
model_comparison_description = """MLP는 lag feature를 2차원 표 형태 입력으로 받아 다음 달 CPI를 예측했다.
LSTM은 과거 6개월의 CPI, 실업률, 금리 흐름을 3차원 입력으로 받아 다음 달 CPI를 예측했다.
두 모델을 같은 검증 구간에서 MAE와 RMSE로 비교해 성능, 복잡도, 해석 가능성을 함께 고려했다."""

print(model_comparison_description)
````

## 7. 평가 설명

````python
evaluation_description = """다음 달 CPI 예측은 숫자를 맞히는 회귀 문제이므로 MAE와 RMSE를 평가 지표로 사용했다.
MAE는 평균적으로 CPI를 얼마나 틀렸는지 보여주며, RMSE는 큰 오차에 더 민감하다.
경제 지표는 외부 충격과 정책 변화의 영향을 받을 수 있으므로 예측 결과를 과장하지 않고 실험 결과로 해석했다."""

print(evaluation_description)
````

## 8. 한계와 개선 방향

````python
limitations = [
    "사용한 경제 지표 수가 제한적이다.",
    "월별 데이터라 샘플 수가 많지 않다.",
    "경제 데이터는 나중에 수정될 수 있다.",
    "정책 변화나 외부 충격을 모델이 충분히 반영하지 못할 수 있다.",
    "LSTM이 항상 MLP보다 좋은 것은 아니다.",
    "예측 결과를 투자 판단이나 정책 판단으로 사용하면 안 된다.",
]

future_work = [
    "ECOS, KOSIS 등 한국 경제 지표를 추가한다.",
    "환율, 유가, 주가지수, 기대인플레이션 등 외부 지표를 추가한다.",
    "전년 동월 대비 변화율 같은 파생 변수를 만든다.",
    "MLP, LSTM, GRU, XGBoost 등 다양한 모델을 비교한다.",
    "walk-forward validation을 적용한다.",
    "Streamlit 대시보드를 제작한다.",
]

print("한계:")
for item in limitations:
    print("-", item)

print("\n개선 방향:")
for item in future_work:
    print("-", item)
````

## 9. README 초안 만들기

````python
readme_text = f"""# {project_info["title"]}

## 1. 프로젝트 개요

{problem_definition}

## 2. 데이터 수집

{data_collection_description}

## 3. 데이터셋 설계

{dataset_design_description}

## 4. 미래 정보 누출 방지

{leakage_prevention_description}

## 5. 모델 비교

{model_comparison_description}

## 6. 평가

{evaluation_description}

## 7. 한계

"""

for item in limitations:
    readme_text += f"- {item}\n"

readme_text += "\n## 8. 개선 방향\n\n"

for item in future_work:
    readme_text += f"- {item}\n"

print(readme_text)
````
