# 24. PM10 시계열 예측 프로젝트 포트폴리오 문서 만들기

이번 장에서는 PM10 시계열 예측 프로젝트를 포트폴리오 문서로 정리합니다.

핵심은 다음 흐름을 설명하는 것입니다.

```text
시계열 데이터 이해
-> 시간 순서 split
-> window 데이터셋
-> baseline 비교
-> LSTM 예측
-> 한계와 개선 방향
```

## 1. 프로젝트 기본 정보

````python
from pathlib import Path

project_info = {
    "title": "환경/건강 데이터 기반 PM10 시계열 예측 프로젝트",
    "dataset": "seoul_pm10.csv",
    "problem_type": "시계열 회귀",
    "target": "다음 시간 PM10",
    "main_model": "LSTM",
    "baselines": ["naive last value", "24-hour mean"],
    "metrics": ["MAE", "RMSE"],
}

project_info
````

## 2. 문제 정의

````python
problem_definition = """서울 지역별 시간 단위 PM10 데이터를 사용해 다음 시간의 PM10 농도를 예측하는 시계열 회귀 모델을 구축한다.
과거 PM10 흐름을 입력으로 사용하고, LSTM 모델과 단순 baseline을 비교해 딥러닝 모델의 실효성을 확인한다.
이 프로젝트는 환경/건강 데이터 기반 시계열 예측 흐름을 학습하고 포트폴리오화하는 것을 목표로 한다."""

print(problem_definition)
````

## 3. 데이터 설명

````python
data_description = """사용 데이터는 서울 지역별 시간 단위 PM10/PM2.5 데이터이다.
주요 컬럼은 date, area, pm10, pm2.5이며, date는 측정 시간, area는 서울 지역구, pm10과 pm2.5는 대기질 지표를 의미한다.
초기 실험에서는 한 지역을 선택해 단일 지역 PM10 시계열 예측 문제로 단순화했다."""

print(data_description)
````

## 4. 전처리 설명

````python
preprocessing_description = """CSV 파일은 cp949 인코딩으로 읽었고, date 컬럼을 datetime 타입으로 변환했다.
지역을 선택한 뒤 시간 순서대로 정렬하고, PM10/PM2.5 결측치는 앞 시간 값으로 채웠다.
미래 정보 누출을 막기 위해 무작위 split을 사용하지 않고, 앞쪽 80%를 학습 데이터, 뒤쪽 20%를 검증 데이터로 사용했다.
MinMaxScaler는 학습 데이터에만 fit하고 검증 데이터에는 transform만 적용했다.
과거 24시간 값을 입력으로, 다음 시간 PM10 값을 정답으로 하는 window 데이터셋을 만들었다."""

print(preprocessing_description)
````

## 5. 모델 설명

````python
model_description = """단변량 LSTM은 과거 24시간의 PM10만 입력으로 사용하며, 입력 shape는 (samples, 24, 1)이다.
다변량 LSTM은 과거 24시간의 PM10과 PM2.5를 함께 입력으로 사용하며, 입력 shape는 (samples, 24, 2)이다.
두 경우 모두 정답은 다음 시간 PM10 하나이므로 출력층은 Dense(1)을 사용했다."""

print(model_description)
````

## 6. Baseline 비교 설명

````python
baseline_description = """LSTM 모델의 성능을 해석하기 위해 naive baseline과 24시간 평균 baseline을 함께 평가했다.
naive baseline은 직전 PM10 값을 다음 값으로 그대로 사용하는 방식이다.
mean baseline은 과거 24시간 PM10 평균을 다음 값으로 사용하는 방식이다.
이를 통해 복잡한 딥러닝 모델이 단순 기준선 대비 실제 개선을 만드는지 확인했다."""

print(baseline_description)
````

## 7. 평가 설명

````python
evaluation_description = """PM10 예측은 숫자를 맞히는 회귀 문제이므로 MAE와 RMSE를 평가 지표로 사용했다.
MAE는 평균적으로 몇 PM10 정도 틀렸는지 보여주고, RMSE는 큰 오차에 더 민감하다.
모델 예측값은 MinMaxScaler로 변환된 값이므로 inverse_transform으로 원래 PM10 단위로 복원한 뒤 평가했다."""

print(evaluation_description)
````

## 8. 한계와 개선 방향

````python
limitations = [
    "초기 실험에서는 한 지역 데이터만 사용했다.",
    "기상 정보, 계절 정보, 요일 정보 등을 사용하지 않았다.",
    "데이터 기간이 제한적이다.",
    "단기 예측에서는 직전 값 baseline이 매우 강할 수 있다.",
    "LSTM이 항상 baseline보다 좋은 것은 아니다.",
]

future_work = [
    "여러 지역을 함께 사용하는 모델로 확장한다.",
    "기온, 습도, 풍속, 강수량 같은 기상 데이터를 추가한다.",
    "시간대, 요일, 계절 특성을 추가한다.",
    "GRU, 1D CNN, Transformer 계열 모델과 비교한다.",
    "1시간 뒤뿐 아니라 6시간 뒤, 24시간 뒤 예측으로 확장한다.",
    "Streamlit 대시보드로 예측 결과를 시각화한다.",
]

print("한계:")
for item in limitations:
    print("-", item)

print("\n개선 방향:")
for item in future_work:
    print("-", item)
````

## 9. 프로젝트 README 초안 생성

````python
readme_text = f"""# {project_info["title"]}

## 1. 프로젝트 개요

{problem_definition}

## 2. 데이터

{data_description}

## 3. 전처리

{preprocessing_description}

## 4. 모델

{model_description}

## 5. Baseline 비교

{baseline_description}

## 6. 평가

{evaluation_description}

## 7. 한계

"""

for item in limitations:
    readme_text += f"- {item}\n"

readme_text += "\n## 8. 개선 방향\n\n"

for item in future_work:
    readme_text += f"- {item}\n"

print(readme_text)
````
