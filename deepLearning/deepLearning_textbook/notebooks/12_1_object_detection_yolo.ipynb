# 12-1. 객체 탐지와 YOLO 입문

이번 노트북은 YOLO를 직접 학습시키는 노트북이 아닙니다.

목표는 이미지 분류와 객체 탐지의 출력 차이를 이해하는 것입니다.

```text
이미지 분류: 이미지 전체에 이름 하나 붙이기
객체 탐지: 이미지 안의 물체 위치와 이름 찾기
```

## 1. 라이브러리 불러오기

````python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np
import pandas as pd
````

## 2. 이미지 분류 출력 예시

````python
# 이미지 분류 모델의 출력은 보통 클래스별 확률입니다.
classification_result = pd.DataFrame([
    {"class": "daisy", "probability": 0.03},
    {"class": "dandelion", "probability": 0.05},
    {"class": "roses", "probability": 0.10},
    {"class": "sunflowers", "probability": 0.07},
    {"class": "tulips", "probability": 0.75},
])

classification_result
````

이미지 분류에서는 가장 높은 확률을 가진 클래스 하나를 최종 예측으로 고릅니다.

````python
top_class = classification_result.sort_values("probability", ascending=False).iloc[0]

print("최종 예측 클래스:", top_class["class"])
print("확률:", top_class["probability"])
````

## 3. 객체 탐지 출력 예시

````python
# 객체 탐지 모델의 출력은 클래스, 위치 상자, 확신도를 함께 가집니다.
detection_result = pd.DataFrame([
    {"class": "person", "confidence": 0.91, "x": 40, "y": 30, "width": 80, "height": 160},
    {"class": "car", "confidence": 0.84, "x": 160, "y": 90, "width": 130, "height": 70},
    {"class": "dog", "confidence": 0.77, "x": 300, "y": 120, "width": 70, "height": 60},
])

detection_result
````

## 4. Bounding Box 그려보기

````python
# 예시용 빈 이미지를 만듭니다.
# 실제 프로젝트에서는 이 위치에 사진이 들어갑니다.
image = np.ones((240, 420, 3))

fig, ax = plt.subplots(figsize=(9, 5))
ax.imshow(image)

for _, row in detection_result.iterrows():
    # Rectangle은 matplotlib에서 사각형을 그리는 도구입니다.
    box = patches.Rectangle(
        (row["x"], row["y"]),
        row["width"],
        row["height"],
        linewidth=2,
        edgecolor="red",
        facecolor="none"
    )
    
    ax.add_patch(box)
    
    label = f"{row['class']} {row['confidence']:.2f}"
    ax.text(row["x"], row["y"] - 5, label, color="red", fontsize=11)

ax.set_title("Object Detection Output Example")
ax.axis("off")
plt.show()
````

## 5. YOLO가 하는 일

YOLO는 이미지를 보고 다음을 동시에 예측합니다.

```text
1. 물체 종류
2. 물체 위치 bounding box
3. confidence
```

````python
yolo_output_columns = pd.DataFrame([
    {"output": "class", "meaning": "탐지한 물체의 종류"},
    {"output": "bounding box", "meaning": "물체를 감싸는 사각형 위치"},
    {"output": "confidence", "meaning": "모델이 이 탐지를 얼마나 확신하는지"},
])

yolo_output_columns
````

## 6. 이미지 분류와 객체 탐지 비교표

````python
comparison = pd.DataFrame([
    {
        "task": "이미지 분류",
        "question": "이미지 전체가 무엇인가?",
        "output": "클래스 하나 또는 클래스별 확률",
        "example": "이 사진은 tulips다",
    },
    {
        "task": "객체 탐지",
        "question": "이미지 안에 무엇이 어디에 있는가?",
        "output": "클래스, 위치 상자, 확신도",
        "example": "이 위치에 person, 저 위치에 car가 있다",
    },
])

comparison
````
