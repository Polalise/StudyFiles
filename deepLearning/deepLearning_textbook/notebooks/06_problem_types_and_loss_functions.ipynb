# 06. 문제 유형과 손실 함수

이번 노트북에서는 문제 유형에 따라 출력층과 손실 함수가 어떻게 달라지는지 확인합니다.

```python
from keras.layers import Dense, Input
from keras.models import Sequential

# 이 노트북은 모델을 오래 학습하지 않습니다.
# 목적은 문제 유형별 모델 마지막 층과 손실 함수 선택을 익히는 것입니다.
```

## 1. 회귀 모델

회귀는 숫자 하나를 예측하는 문제입니다.

예: 집값, 기온, PM10 수치, 자전거 대여량

```python
regression_model = Sequential([
    Input(shape=(3,)),  # 입력 특성이 3개라고 가정합니다.
    Dense(8, activation="relu"),  # 중간층입니다. 복잡한 패턴을 배우기 위한 층입니다.
    Dense(1)  # 회귀는 숫자 하나를 예측하므로 출력이 1개입니다.
])

regression_model.compile(
    optimizer="adam",  # adam은 자주 쓰는 최적화 방법입니다.
    loss="mse"  # 회귀에서는 평균 제곱 오차를 자주 사용합니다.
)

regression_model.summary()
```

## 2. 이진 분류 모델

이진 분류는 둘 중 하나를 고르는 문제입니다.

예: 당뇨 위험 있음/없음, 스팸/정상, 심장 질환 있음/없음

```python
binary_model = Sequential([
    Input(shape=(10,)),  # 입력 특성이 10개라고 가정합니다.
    Dense(16, activation="relu"),  # 은닉층에서는 ReLU를 자주 사용합니다.
    Dense(1, activation="sigmoid")  # 이진 분류는 0~1 사이 확률 하나를 출력합니다.
])

binary_model.compile(
    optimizer="adam",
    loss="binary_crossentropy",  # 이진 분류에 자주 쓰는 손실 함수입니다.
    metrics=["accuracy"]  # accuracy는 맞힌 비율입니다.
)

binary_model.summary()
```

## 3. 다중 분류 모델

다중 분류는 여러 종류 중 하나를 고르는 문제입니다.

예: 꽃 5종류 분류, 숫자 0~9 분류

```python
class_count = 5  # 꽃 종류가 5개라고 가정합니다.

multi_model = Sequential([
    Input(shape=(20,)),  # 입력 특성이 20개라고 가정합니다.
    Dense(32, activation="relu"),
    Dense(class_count, activation="softmax")  # 클래스 수만큼 출력하고 softmax로 확률처럼 만듭니다.
])

multi_model.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",  # 정답이 0,1,2,3,4 같은 정수 라벨일 때 사용합니다.
    metrics=["accuracy"]
)

multi_model.summary()
```
