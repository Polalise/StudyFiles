## 2. 프로젝트 후보 점수표 만들기

````python
project_candidates = pd.DataFrame([
    {
        "project": "당뇨병 위험 예측",
        "portfolio_strength": 5,
        "difficulty": 3,
        "demo_easy": 5,
        "explainability": 5,
        "note": "대표 프로젝트 1순위 후보",
    },
    {
        "project": "PM10 시계열 예측",
        "portfolio_strength": 4,
        "difficulty": 4,
        "demo_easy": 4,
        "explainability": 4,
        "note": "환경/건강 시계열 프로젝트",
    },
    {
        "project": "꽃 이미지 분류",
        "portfolio_strength": 4,
        "difficulty": 3,
        "demo_easy": 5,
        "explainability": 3,
        "note": "시각적 데모가 강함",
    },
    {
        "project": "경제 시계열 예측",
        "portfolio_strength": 4,
        "difficulty": 5,
        "demo_easy": 3,
        "explainability": 3,
        "note": "API 수집과 시계열 설계가 강점",
    },
])

project_candidates
````

## 3. 대표 프로젝트 추천 순위 계산

````python
# 간단한 추천 점수입니다.
# 포트폴리오 강점, 데모 용이성, 설명 가능성은 높을수록 좋고,
# 난이도는 너무 높으면 부담이 되므로 감점합니다.
project_candidates["recommend_score"] = (
    project_candidates["portfolio_strength"] * 2
    + project_candidates["demo_easy"]
    + project_candidates["explainability"]
    - project_candidates["difficulty"]
)

project_candidates.sort_values("recommend_score", ascending=False)
````
