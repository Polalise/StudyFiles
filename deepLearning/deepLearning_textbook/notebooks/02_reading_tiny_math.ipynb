# 02. 아주 작은 수식 읽기

이번 노트북에서는 `y = wx + b`를 코드로 직접 계산해봅니다.

목표는 수식을 잘 푸는 것이 아니라, 수식을 문장처럼 읽는 감각을 만드는 것입니다.

## 1. 라이브러리 불러오기

이번 장에서는 딥러닝 라이브러리를 쓰지 않습니다.

먼저 파이썬과 `numpy`만으로 수식이 코드에서 어떻게 움직이는지 확인합니다.

````python
import numpy as np  # 여러 숫자를 배열로 묶고 계산하기 위한 라이브러리입니다.
import matplotlib.pyplot as plt  # w와 b가 바뀔 때 그래프가 어떻게 달라지는지 보기 위한 라이브러리입니다.
````

## 2. y = wx + b를 함수로 만들기

수식:

```text
y = wx + b
```

파이썬 코드:

```python
y = w * x + b
```

수식에서는 `wx`처럼 곱하기 기호를 생략할 수 있지만, 파이썬에서는 반드시 `*`를 써야 합니다.

````python
def predict(x, w, b):
    # 이 함수는 y = wx + b를 코드로 적은 것입니다.
    # x: 입력값
    # w: 가중치, 입력값에 곱해지는 숫자
    # b: 편향, 기본으로 더해지는 숫자
    # return은 계산한 예측값 y를 함수 밖으로 돌려주는 명령입니다.
    return w * x + b
````

## 3. 숫자 하나 넣어보기

먼저 `y = 3x + 2`를 계산해봅니다.

이 수식은 이렇게 읽습니다.

```text
입력값 x에 3을 곱하고 2를 더하면 y가 된다.
```

````python
x = 4  # 입력값입니다.
w = 3  # 입력값에 곱할 가중치입니다.
b = 2  # 마지막에 더할 편향입니다.

y = predict(x, w, b)

print("입력 x:", x)
print("가중치 w:", w)
print("편향 b:", b)
print("예측 y:", y)
````

## 4. 여러 입력을 한 번에 계산하기

`numpy` 배열을 쓰면 여러 입력값을 한 번에 계산할 수 있습니다.

이것은 딥러닝에서 매우 중요합니다. 모델은 보통 데이터 한 개만 보는 것이 아니라 여러 샘플을 한 번에 봅니다.

````python
# np.array()는 여러 숫자를 하나의 배열로 묶어줍니다.
# 여기서는 x가 0, 1, 2, 3, 4일 때의 y를 한 번에 계산합니다.
x_values = np.array([0, 1, 2, 3, 4], dtype=np.float32)

w = 3
b = 2

y_values = predict(x_values, w, b)

print("입력 x:", x_values)
print("예측 y:", y_values)
````

## 5. w가 바뀌면 무엇이 달라질까

`w`는 입력값에 주는 힘입니다.

같은 x라도 `w`가 커지면 y가 더 빠르게 커집니다.

````python
x_values = np.array([0, 1, 2, 3, 4], dtype=np.float32)
b = 0

for w in [1, 2, 5]:
    # w 값만 바꾸면서 y가 어떻게 달라지는지 확인합니다.
    y_values = predict(x_values, w, b)
    print(f"w={w}, b={b}일 때 y:", y_values)
````

````python
# 그래프로도 확인합니다.
# w가 클수록 선이 더 가파르게 올라갑니다.
x_values = np.array([0, 1, 2, 3, 4], dtype=np.float32)
b = 0

for w in [1, 2, 5]:
    y_values = predict(x_values, w, b)
    plt.plot(x_values, y_values, marker="o", label=f"w={w}, b={b}")

plt.title("Effect of w")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
````

## 6. b가 바뀌면 무엇이 달라질까

`b`는 기본으로 더해지는 값입니다.

그래프에서는 선 전체를 위아래로 옮기는 역할을 합니다.

````python
x_values = np.array([0, 1, 2, 3, 4], dtype=np.float32)
w = 2

for b in [0, 3, -2]:
    # b 값만 바꾸면서 y가 어떻게 달라지는지 확인합니다.
    y_values = predict(x_values, w, b)
    print(f"w={w}, b={b}일 때 y:", y_values)
````

````python
# 그래프로도 확인합니다.
# b가 바뀌면 선의 기울기는 같고, 위치만 위아래로 움직입니다.
x_values = np.array([0, 1, 2, 3, 4], dtype=np.float32)
w = 2

for b in [0, 3, -2]:
    y_values = predict(x_values, w, b)
    plt.plot(x_values, y_values, marker="o", label=f"w={w}, b={b}")

plt.title("Effect of b")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
````

## 7. 정리

이번 장에서 기억할 것은 세 가지입니다.

```text
x는 입력값입니다.
w는 입력값에 곱해지는 가중치입니다.
b는 기본으로 더해지는 편향입니다.
```

그리고 수식은 이렇게 코드로 바뀝니다.

```text
수식: y = wx + b
코드: y = w * x + b
```
