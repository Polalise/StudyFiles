17장에서는 예측 함수를 만들었고, 18장에서는 포트폴리오 문서를 만들었다.

이번 노트북에서는 그 결과를 작은 데모 앱으로 연결하는 방법을 배운다.

이 노트북은 Streamlit 서버를 직접 실행하지 않는다.
대신 앱 파일을 생성하고, 실행 명령과 문법 검사를 확인한다.


````python

from pathlib import Path
import py_compile

project_dir = Path(r"C:\work\deepLearning\deepLearning_textbook")
app_path = project_dir / "streamlit_diabetes_demo_app.py"

print("앱 파일 경로:", app_path)
````

## Streamlit 앱의 기본 흐름

Streamlit 앱은 보통 `app.py` 같은 Python 파일로 만든다.

앱 안에서는 다음 일을 한다.

```text
1. 제목과 설명을 표시한다.
2. 사용자가 입력할 위젯을 만든다.
3. 버튼을 누르면 예측 코드를 실행한다.
4. 예측 점수와 주의 문구를 표시한다.
```

실제 프로젝트에서는 17장에서 만든 예측 함수를 버튼 안에서 호출하면 된다.


````python

app_code = 'import streamlit as st\nimport pandas as pd\n\n\nst.set_page_config(\n    page_title="Diabetes Risk Demo",\n    page_icon="",\n    layout="centered",\n)\n\nst.title("당뇨병 위험 예측 데모")\n\nst.caption(\n    "이 앱은 학습용 포트폴리오 데모입니다. "\n    "의학적 진단이나 치료 판단에 사용할 수 없습니다."\n)\n\nst.subheader("입력값")\n\n# 실제 프로젝트에서는 학습 때 사용한 feature_names와 같은 순서를 사용해야 한다.\n# 여기서는 데모 구조를 보여주기 위해 일부 대표 입력만 사용한다.\nage = st.slider("Age", min_value=18, max_value=100, value=45)\nbmi = st.number_input("BMI", min_value=10.0, max_value=60.0, value=25.0, step=0.1)\nhigh_bp = st.selectbox("High Blood Pressure", options=[0, 1], format_func=lambda x: "Yes" if x == 1 else "No")\nhigh_chol = st.selectbox("High Cholesterol", options=[0, 1], format_func=lambda x: "Yes" if x == 1 else "No")\nphysical_activity = st.selectbox("Physical Activity", options=[0, 1], format_func=lambda x: "Yes" if x == 1 else "No")\n\ninput_df = pd.DataFrame([{\n    "Age": age,\n    "BMI": bmi,\n    "HighBP": high_bp,\n    "HighChol": high_chol,\n    "PhysActivity": physical_activity,\n}])\n\nst.write("입력 확인")\nst.dataframe(input_df, use_container_width=True)\n\nif st.button("예측하기"):\n    # 실제 프로젝트에서는 이 부분을 17장에서 만든 예측 함수로 바꾼다.\n    #\n    # 예시:\n    # model = keras.models.load_model("diabetes_model.keras")\n    # scaler = joblib.load("diabetes_scaler.joblib")\n    # feature_names = joblib.load("feature_names.joblib")\n    # score = predict_diabetes_risk(input_df, model, scaler, feature_names)\n    #\n    # 지금은 앱 구조를 보여주기 위한 임시 점수 계산식이다.\n    risk_score = (\n        0.15\n        + 0.004 * (age - 18)\n        + 0.018 * max(bmi - 20, 0)\n        + 0.14 * high_bp\n        + 0.10 * high_chol\n        - 0.06 * physical_activity\n    )\n    risk_score = min(max(risk_score, 0.0), 1.0)\n\n    threshold = 0.50\n    predicted_class = int(risk_score >= threshold)\n\n    st.metric("예측 점수", f"{risk_score:.2f}")\n\n    if predicted_class == 1:\n        st.warning("모델 기준상 위험 점수가 높게 계산되었습니다.")\n    else:\n        st.success("모델 기준상 위험 점수가 낮게 계산되었습니다.")\n\n    st.info(\n        "이 결과는 학습용 예시입니다. "\n        "실제 의료 판단은 전문가 상담과 검사를 통해 이루어져야 합니다."\n    )\n\nst.divider()\nst.write("포트폴리오 설명 예시")\nst.markdown(\n    "- 노트북에서 학습한 모델을 사용자 입력 기반 데모로 확장하는 흐름을 보여줍니다.\\n"\n    "- 실제 배포 시에는 저장된 모델, scaler, feature order를 함께 불러와야 합니다.\\n"\n    "- 건강 관련 결과는 진단이 아니라 예측 점수로 표현해야 합니다."\n)\n'

app_path.write_text(app_code, encoding="utf-8")
print("Streamlit 앱 파일을 생성했습니다:")
print(app_path)
````

## 생성된 앱 파일 확인하기

아래 셀은 생성된 Python 파일의 앞부분을 출력한다.
파일이 제대로 만들어졌는지 빠르게 확인하기 위한 단계다.


````python

preview_lines = app_path.read_text(encoding="utf-8").splitlines()[:30]
for line_no, line in enumerate(preview_lines, start=1):
    print(f"{line_no:02d}: {line}")
````

## 문법 검사하기

Streamlit 앱을 실행하기 전에 Python 문법 오류가 없는지 확인한다.

`py_compile.compile()`은 파일을 실행하지 않고 문법상 문제가 있는지 검사한다.


````python

py_compile.compile(str(app_path), doraise=True)
print("문법 검사 통과:", app_path.name)
````

## 실행 명령

실제로 앱 화면을 보고 싶다면 VS Code 터미널에서 아래 명령을 실행한다.

노트북 셀에서 직접 서버를 켜면 실행이 계속 대기할 수 있으므로, 터미널 실행을 권장한다.


````python

command = f"streamlit run {app_path}"
print(command)
````

## 실제 모델과 연결하려면

지금 생성한 앱은 구조를 보여주기 위해 임시 위험 점수 계산식을 사용한다.

실제 프로젝트에서는 버튼 안의 임시 계산식을 17장의 예측 함수로 바꾸어야 한다.

필요한 것은 다음 세 가지다.

