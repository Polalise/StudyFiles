## 0.23 이 장의 핵심 정리

이 장은 실습을 위한 준비 단계입니다. 이후 장에서 다룰 핵심 흐름은 다음과 같이 이어집니다.

1. NLP: 텍스트를 토큰화, 인코딩, 패딩, 임베딩합니다.
2. Transformer: Attention, Q/K/V, Encoder/Decoder 구조를 이해하고 직접 구현합니다.
3. Hugging Face: Pipeline으로 사전학습 모델을 쉽게 사용하고, 모델을 Hub에 업로드합니다.
4. Ollama: 로컬 LLM을 실행하고 LangChain과 연결합니다.
5. LangChain: Prompt, Chain, Memory, Retriever를 조합합니다.
6. RAG: 외부 문서를 검색해 LLM 답변의 근거로 사용합니다.
7. Gradio: 실습 결과를 웹 UI로 제공합니다.
8. 확장 실습: STT/TTS, 이미지 분류, DB 질의, 자기소개서 생성으로 LLM 활용 범위를 넓힙니다.
