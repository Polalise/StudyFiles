## 21.4 기본 LLM 챗봇 프로젝트

가장 기본적인 프로젝트는 사용자의 텍스트 질문에 LLM이 답하는 챗봇입니다.

관련 장은 다음과 같습니다.

- 03장 Hugging Face Pipeline
- 04장 로컬 LLM 실행
- 05장 LangChain 기본
- 06장 Gradio 기반 LLM 챗봇

기본 구조는 다음과 같습니다.

```text
사용자 질문
  -> PromptTemplate
  -> LLM
  -> OutputParser
  -> Gradio UI 출력
```

이 구조는 이후 모든 프로젝트의 출발점입니다.


````python
# 교재 위치: 21장 전체 실습 프로젝트 정리 - 기본 챗봇 구조 예시
# 이 셀은 전체 교재의 가장 기본이 되는 LLM 챗봇 구조를 요약 코드로 보여줍니다.

# ChatPromptTemplate은 채팅 모델에 전달할 프롬프트를 구성합니다.
from langchain_core.prompts import ChatPromptTemplate

# StrOutputParser는 LLM 응답을 문자열로 변환합니다.
from langchain_core.output_parsers import StrOutputParser

# ChatOllama는 Ollama 로컬 LLM을 LangChain 방식으로 호출합니다.
from langchain_ollama import ChatOllama

# llm 변수에는 사용할 로컬 모델 이름을 지정합니다.
llm = ChatOllama(model="llama3.1")

# basic_prompt는 기본 챗봇용 프롬프트입니다.
basic_prompt = ChatPromptTemplate.from_messages([
    # system 메시지는 챗봇의 역할을 지정합니다.
    ("system", "너는 LLM 기초 실습을 도와주는 한국어 튜터야."),
    # human 메시지는 사용자의 질문을 받습니다.
    ("human", "질문: {question}"),
])

# basic_chain은 프롬프트, LLM, 출력 파서를 연결한 기본 체인입니다.
basic_chain = basic_prompt | llm | StrOutputParser()

# ask_basic_chatbot 함수는 질문을 받아 기본 체인의 답변을 반환합니다.
def ask_basic_chatbot(question: str) -> str:
    # 질문이 비어 있으면 안내 문장을 반환합니다.
    if not question:
        # 빈 질문에 대한 안내입니다.
        return "질문을 입력해 주세요."

    # basic_chain을 실행해 답변을 생성합니다.
    answer = basic_chain.invoke({"question": question})

    # 생성된 답변을 반환합니다.
    return answer
````

## 21.5 RAG 챗봇 프로젝트

RAG 챗봇은 사용자의 질문에 답하기 전에 관련 문서를 검색합니다.

관련 장은 다음과 같습니다.

- 07장 RAG 검색증강생성 기본
- 08장 FAISS 기반 RAG 실습
- 09장 Chroma 기반 RAG 실습
- 10장 웹 URL 기반 RAG와 웹 문서 요약
- 11장 벡터스토어와 데이터베이스 비교

기본 구조는 다음과 같습니다.

```text
문서 로드
  -> 문서 분할
  -> 임베딩
  -> 벡터스토어 저장
  -> 질문 입력
  -> 관련 문서 검색
  -> LLM 답변 생성
```

RAG에서 가장 중요한 것은 검색 품질입니다. LLM이 아무리 좋아도 잘못된 문서를 근거로 받으면 답변 품질이 낮아질 수 있습니다.


````python
# 교재 위치: 21장 전체 실습 프로젝트 정리 - RAG 구조 요약 함수
# 이 셀은 RAG 프로젝트의 실행 단계를 문자열로 정리하는 예시입니다.

# describe_rag_pipeline 함수는 RAG 실행 흐름을 단계별로 반환합니다.
def describe_rag_pipeline() -> str:
    # steps 리스트에는 RAG 실행 단계를 순서대로 저장합니다.
    steps = [
        # 첫 번째 단계는 원본 문서를 읽는 것입니다.
        "1. PDF, TXT, 웹 페이지 같은 원본 문서를 로드합니다.",
        # 두 번째 단계는 문서를 청크로 나누는 것입니다.
        "2. TextSplitter로 문서를 작은 청크로 나눕니다.",
        # 세 번째 단계는 청크를 임베딩 벡터로 변환하는 것입니다.
        "3. Embedding 모델로 각 청크를 벡터로 변환합니다.",
        # 네 번째 단계는 벡터스토어에 저장하는 것입니다.
        "4. FAISS 또는 Chroma 같은 벡터스토어에 저장합니다.",
        # 다섯 번째 단계는 질문과 관련된 문서를 검색하는 것입니다.
        "5. Retriever로 사용자 질문과 관련된 문서를 검색합니다.",
        # 여섯 번째 단계는 검색 결과를 프롬프트에 넣는 것입니다.
        "6. 검색 결과를 context로 정리해 프롬프트에 넣습니다.",
        # 일곱 번째 단계는 LLM이 답변을 생성하는 것입니다.
        "7. LLM이 context를 근거로 최종 답변을 생성합니다.",
    ]

    # 줄바꿈으로 단계를 연결해 하나의 문자열로 반환합니다.
    return "\n".join(steps)


# RAG 파이프라인 설명을 출력합니다.
print(describe_rag_pipeline())
````

## 21.6 DB 접속 챗봇 프로젝트

DB 접속 챗봇은 문서 검색이 아니라 정형 데이터 조회를 중심으로 합니다.

관련 장은 다음과 같습니다.

- 11장 벡터스토어와 데이터베이스 비교
- 12장 DB 접속 LLM 챗봇
- 13장 LangGraph의 DB 분기 예시
- 14장 LangSmith의 DB 추적 예시

기본 구조는 다음과 같습니다.

```text
사용자 자연어 질문
  -> DB 스키마 확인
  -> LLM으로 SQL 생성
  -> SQL 안전성 검증
  -> SQL 실행
  -> 결과를 자연어로 설명
```

DB 챗봇에서는 보안이 특히 중요합니다. 실습에서도 `SELECT` 중심의 읽기 전용 쿼리만 허용하는 구조를 권장합니다.


````python
# 교재 위치: 21장 전체 실습 프로젝트 정리 - SQL 안전성 검사 요약
# 이 셀은 DB 챗봇에서 반복적으로 사용하는 SQL 안전성 검사 함수를 요약합니다.

# is_safe_select_sql 함수는 SQL이 읽기 전용 SELECT 문인지 확인합니다.
def is_safe_select_sql(sql: str) -> bool:
    # SQL이 비어 있으면 안전하지 않다고 판단합니다.
    if not sql:
        # 비어 있는 SQL은 실행할 수 없으므로 False입니다.
        return False

    # SQL을 소문자로 바꿔 키워드 검사를 쉽게 합니다.
    lowered_sql = sql.lower()

    # 데이터 변경이나 삭제를 일으킬 수 있는 키워드를 목록으로 정의합니다.
    blocked_keywords = ["insert", "update", "delete", "drop", "alter", "truncate", "create"]

    # 위험 키워드가 하나라도 포함되어 있는지 확인합니다.
    has_blocked_keyword = any(keyword in lowered_sql for keyword in blocked_keywords)

    # SELECT로 시작하고 위험 키워드가 없으면 안전한 SQL로 판단합니다.
    return lowered_sql.strip().startswith("select") and not has_blocked_keyword


# 안전한 SELECT 예시입니다.
safe_sql_example = "SELECT customer_id, COUNT(*) FROM orders GROUP BY customer_id;"

# 위험한 DELETE 예시입니다.
unsafe_sql_example = "DELETE FROM orders;"

# 안전한 SQL 검사 결과를 출력합니다.
print("safe_sql_example:", is_safe_select_sql(safe_sql_example))

# 위험한 SQL 검사 결과를 출력합니다.
print("unsafe_sql_example:", is_safe_select_sql(unsafe_sql_example))
````

## 21.7 LangGraph 기반 고도화

LangGraph는 복잡한 LLM 앱을 상태 기반 그래프로 관리합니다.

적용할 수 있는 구조는 다음과 같습니다.

| 프로젝트 | LangGraph 적용 방식 |
|---|---|
| 기본 챗봇 | 질문 분류 후 일반 답변 또는 RAG로 분기 |
| RAG 챗봇 | 검색, 문서 정리, 답변 생성, 근거 검증을 노드화 |
| DB 챗봇 | SQL 생성, 검증, 실행, 설명을 노드화 |
| 자기소개서 챗봇 | 피드백, 개선, 면접 질문 생성을 노드화 |
| 다중 에이전트 | Planner, Writer, Reviewer를 노드화 |

LangGraph를 도입하면 각 단계의 입력과 출력을 명확히 관리할 수 있고, 조건 분기와 반복 흐름을 표현하기 쉬워집니다.


## 21.8 LangSmith 기반 추적과 평가

LLM 앱은 실행 결과가 매번 달라질 수 있으므로 추적과 평가가 중요합니다.

LangSmith를 사용하면 다음을 확인할 수 있습니다.

- 어떤 프롬프트가 LLM에 전달되었는가
- 어떤 문서가 RAG에서 검색되었는가
- SQL 생성과 검증 과정이 어떻게 진행되었는가
- LangGraph의 각 노드가 어떤 State를 만들었는가
- 실패 사례가 어떤 입력에서 발생했는가
- 개선 전후의 답변 품질이 어떻게 달라졌는가

프로젝트를 완성한 뒤에는 단순히 한 번 실행해 보는 것보다, 평가 질문 목록을 만들고 반복 실행해 품질을 비교하는 것이 좋습니다.
