## 1.1 LLM이란?

LLM은 `Large Language Model`, 즉 거대언어모델입니다. 대규모 텍스트 데이터를 학습해 사람의 언어를 이해하고 생성하는 인공지능 모델을 의미합니다.

PDF 교재에서는 LLM의 특징을 다음 네 가지로 설명합니다.

| 특징 | 설명 |
|---|---|
| 대규모 데이터 학습 | 인터넷에 공개된 웹페이지, 도서, 논문 등 다양한 자료를 바탕으로 학습합니다. |
| 문맥 이해 | 단순히 단어를 나열하지 않고, 문맥과 의미를 고려해 결과를 생성합니다. |
| 다양한 작업 수행 | 질의응답, 글쓰기, 번역, 요약, 코딩, 추천 시스템 등 폭넓게 활용됩니다. |
| 사전학습과 미세조정 | 먼저 일반 언어 이해 능력을 익히고, 이후 특정 작업에 맞게 Fine-tuning할 수 있습니다. |

즉 LLM은 단순한 문장 자동완성기가 아니라, 대량의 언어 패턴을 학습한 뒤 입력 문맥에 맞춰 답변, 요약, 번역, 코드, 추천 등을 생성하는 범용 언어 모델입니다.


## 1.2 사전학습과 미세조정

PDF 참고자료 9.1에서는 사전학습과 파인튜닝을 다음처럼 설명합니다.

### 사전학습, Pre-training

LLM이 가장 먼저 수행하는 학습입니다. 인터넷상의 방대한 텍스트 데이터를 읽고 다음에 올 단어를 예측하는 훈련을 반복합니다.

- 막대한 비용과 시간이 필요합니다.
- 일반 언어 능력과 상식은 넓게 얻지만, 특정 분야 전문성은 부족할 수 있습니다.
- 비유하면 초등학생부터 대학생까지 기초 교양과 상식을 공부해 문해력을 갖추는 단계입니다.

### 미세조정, Fine-tuning

사전학습이 끝난 모델에 특정 목적에 맞는 데이터를 추가 학습시켜 전문화하는 과정입니다.

- 의료, 법률, 고객센터, 기업 내부 문서 등 특정 데이터로 모델을 조정할 수 있습니다.
- 사전학습보다 적은 데이터와 자원으로 가능합니다.
- 모델의 말투, 답변 형식, 특정 도메인 지식을 조정할 수 있습니다.
- 비유하면 대학 졸업 후 의대나 법학전문대학원에 진학해 특정 직무 훈련을 받는 것입니다.


## 1.3 NLP, 자연어처리란?

NLP는 `Natural Language Processing`, 즉 자연어처리입니다. 사람이 일상적으로 사용하는 언어를 컴퓨터가 이해하고 처리할 수 있도록 만드는 기술입니다.

PDF 교재의 NLP 특징은 다음과 같습니다.

| 특징 | 설명 |
|---|---|
| 인간 언어의 복잡성 | 다양한 표현 방식, 문법 구조, 모호성을 처리해야 합니다. |
| 비구조적 데이터 처리 | 텍스트나 음성과 같은 자유로운 형식의 데이터를 구조화하여 분석합니다. |
| 기계학습과 딥러닝 활용 | 머신러닝, 딥러닝 모델을 기반으로 언어 데이터를 학습하고 처리합니다. |
| 언어별 특화 접근 필요 | 언어마다 문법과 특성이 달라 맞춤형 처리가 필요합니다. |
| 문맥 및 의미 이해 | 단어만 처리하지 않고 문맥과 의미를 파악해야 합니다. |
| 주요 작업 구성 | 텍스트 전처리, 토큰화, 품사 태깅, 문맥 이해 등을 포함합니다. |

LLM을 사용하더라도 입력은 결국 텍스트이고, 모델 내부에서는 텍스트가 숫자와 벡터로 변환됩니다. 그래서 NLP의 기본 처리 과정을 이해해야 이후 Transformer와 RAG를 제대로 이해할 수 있습니다.


## 1.4 NLP 처리 과정 전체 흐름

PDF 교재의 NLP 과정은 다음 흐름으로 정리할 수 있습니다.

```text
원문 텍스트
  -> 토크나이징(Tokenizing)
  -> 인코딩(Encoding)
  -> 패딩(Padding)
  -> 임베딩(Embedding)
  -> 모델 입력
```

각 단계의 역할은 다음과 같습니다.

| 단계 | 역할 | 예시 |
|---|---|---|
| 토크나이징 | 문장을 작은 단위로 나눕니다. | `한국어 NLP는 재미있어요` → `한국어`, `NLP`, `는`, `재미있어요` |
| 인코딩 | 토큰을 숫자 ID로 바꿉니다. | `한국어` → 1, `NLP` → 2 |
| 패딩 | 문장 길이를 같게 맞춥니다. | `[1, 2, 3]` → `[1, 2, 3, 0, 0]` |
| 임베딩 | 숫자 ID를 의미 벡터로 바꿉니다. | `1` → `[0.12, -0.03, ...]` |

PDF의 RAG 핵심기술 장에서도 자연어처리 과정, 토크나이저, 임베딩이 다시 등장합니다. RAG에서는 문서와 질문을 모두 임베딩 벡터로 바꾸고, 벡터 간 유사도를 비교해 관련 문서를 찾기 때문입니다.


## 1.5 토크나이징, Tokenizing

토크나이징은 문장을 모델이 처리할 수 있는 작은 단위로 나누는 과정입니다. 이 작은 단위를 토큰이라고 합니다.

토큰은 반드시 단어 하나와 일치하지 않습니다.

- 영어에서는 공백 기준 단어가 토큰이 되는 경우가 많습니다.
- 한국어에서는 조사, 어미, 형태소 때문에 단순 공백 분리만으로는 부족합니다.
- BERT, GPT 같은 모델은 단어보다 더 작은 subword 단위를 사용하기도 합니다.
- 문자 단위, 형태소 단위, 단어 단위, subword 단위 등 다양한 방식이 있습니다.

예를 들어 `자연어 처리는 재미있는 분야입니다.`라는 문장은 형태소 분석기를 거치면 `자연어`, `처리`, `는`, `재미있는`, `분야`, `입니다`, `.`처럼 나뉠 수 있습니다.


## 1.6 인코딩, Encoding

인코딩은 토큰을 숫자로 바꾸는 과정입니다. 컴퓨터와 딥러닝 모델은 문자열 자체를 직접 계산하지 못하므로, 각 토큰에 정수 ID를 부여합니다.

예시:

```text
단어 사전
<PAD>  -> 0
자연어 -> 1
처리   -> 2
재미   -> 3

문장: 자연어 처리
인코딩 결과: [1, 2]
```

이렇게 만들어진 숫자 ID 배열을 모델의 입력으로 사용할 수 있습니다. 다만 숫자 ID 자체에는 의미적 거리가 없습니다. 예를 들어 ID 1과 ID 2가 가깝다고 해서 두 단어의 의미가 비슷하다는 뜻은 아닙니다. 의미적 관계는 다음 단계인 임베딩에서 학습됩니다.


## 1.7 패딩, Padding

딥러닝 모델은 보통 같은 크기의 입력 묶음, 즉 batch를 처리합니다. 그런데 문장은 길이가 서로 다릅니다.

예시:

```text
문장 A: [1, 2, 3]
문장 B: [4, 5]
문장 C: [6, 7, 8, 9, 10]
```

이 문장들을 한 번에 모델에 넣으려면 길이를 맞춰야 합니다. 부족한 부분은 보통 0으로 채웁니다.

```text
문장 A: [1, 2, 3, 0, 0]
문장 B: [4, 5, 0, 0, 0]
문장 C: [6, 7, 8, 9, 10]
```

여기서 0은 실제 단어가 아니라 `<PAD>` 같은 특수 토큰으로 취급됩니다.


## 1.8 임베딩, Embedding

임베딩은 토큰 ID를 의미가 담긴 벡터로 바꾸는 과정입니다.

PDF 교재에서는 임베딩을 다음처럼 설명합니다.

- 숫자로 변환된 단어를 고차원의 실수 벡터로 변환합니다.
- 벡터 간 거리는 의미적 유사성을 나타냅니다.
- 예를 들어 의미가 비슷한 단어는 벡터 공간에서 가까운 위치에 놓입니다.

예시:

```text
사과 -> [0.104, 0.248, -0.531, ...]
바나나 -> [0.099, 0.231, -0.488, ...]
자동차 -> [-0.721, 0.012, 0.334, ...]
```

`사과`와 `바나나`는 과일이라는 의미가 비슷하므로 벡터도 가까울 가능성이 높습니다. 반면 `자동차`는 의미가 달라 멀리 떨어질 수 있습니다.

RAG에서는 문서 청크와 사용자 질문을 임베딩으로 변환한 뒤, 코사인 유사도 같은 방식으로 가장 관련 있는 청크를 찾습니다.

````python
# 필요한 패키지 설치 예시입니다.
# 환경에 이미 설치되어 있다면 실행하지 않아도 됩니다.

# %pip install konlpy
# %pip install JPype1==1.4.1
# %pip install numpy tensorflow
````

````python
# KoNLPy의 Okt는 한국어 형태소 분석을 수행하는 도구입니다.
# numpy는 수치 계산을 위한 라이브러리입니다.
# Keras Tokenizer는 텍스트를 단어 인덱스로 바꾸는 데 사용합니다.
# pad_sequences는 길이가 다른 문장 시퀀스를 같은 길이로 맞춥니다.
# Embedding은 토큰 ID를 벡터로 바꾸는 신경망 층입니다.
# Sequential은 Keras 모델을 순서대로 쌓아 만드는 방식입니다.

from konlpy.tag import Okt
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.layers import Embedding
from tensorflow.keras.models import Sequential
````

## 1.10 예제 문장 준비

원본 노트북은 한국어 문장 세 개를 사용합니다.

```text
자연어 처리는 재미있는 분야입니다.
딥러닝은 많은 데이터를 필요로 합니다.
한국어 NLP는 정말 재미있어요!
```

이 문장들은 토크나이징, 인코딩, 패딩, 임베딩의 전체 과정을 보여주기 위해 사용됩니다.


````python
# 1. 텍스트 데이터입니다.
# 각 문자열은 NLP 처리 과정의 입력 문장입니다.
sentences = [
    "자연어 처리는 재미있는 분야입니다.",
    "딥러닝은 많은 데이터를 필요로 합니다.",
    "한국어 NLP는 정말 재미있어요!"
]

# 준비한 문장 목록을 확인합니다.
sentences
````

## 1.11 한국어 토크나이징 실습

한국어는 조사와 어미가 단어 뒤에 붙는 경우가 많기 때문에 영어처럼 공백만으로 나누면 의미 단위를 잘못 잡을 수 있습니다.

예를 들어 `딥러닝은`은 `딥러닝`과 `은`으로 나누는 것이 후속 분석에 더 유리할 수 있습니다.

여기서는 KoNLPy의 `Okt().morphs()`를 사용해 문장을 형태소 단위로 나눕니다.


````python
# 2. 토크나이징 단계입니다.
# Okt 객체를 생성합니다.
okt = Okt()

# 각 문장을 형태소 단위의 토큰 목록으로 변환합니다.
tokenized_sentences = [okt.morphs(sentence) for sentence in sentences]

# 토크나이징 결과를 출력합니다.
print("토크나이징 결과:", tokenized_sentences)
````

## 1.12 인코딩 실습

토큰화된 문장을 모델에 입력하려면 각 토큰을 숫자 ID로 바꿔야 합니다.

Keras의 `Tokenizer`는 다음 일을 합니다.

1. 전체 토큰 목록을 보고 단어 사전을 만듭니다.
2. 각 토큰에 정수 인덱스를 부여합니다.
3. 문장을 정수 시퀀스로 변환합니다.

`tokenizer.word_index`를 보면 어떤 토큰이 어떤 숫자로 매핑되었는지 확인할 수 있습니다.


````python
# 3. 인코딩 단계입니다.
# Tokenizer 객체를 생성합니다.
tokenizer = Tokenizer()

# 토큰화된 문장 목록을 기준으로 단어 사전을 학습합니다.
tokenizer.fit_on_texts(tokenized_sentences)

# 각 문장을 숫자 ID 시퀀스로 변환합니다.
encoded_sentences = tokenizer.texts_to_sequences(tokenized_sentences)

# 인코딩 결과를 출력합니다.
print("인코딩 결과:", encoded_sentences)

# 단어와 숫자 ID의 매핑을 확인합니다.
print("단어 사전:", tokenizer.word_index)
````

## 1.13 패딩 실습

문장마다 토큰 개수가 다르므로, 모델에 batch로 입력하려면 길이를 맞춰야 합니다.

원본 노트북은 최대 길이를 `10`으로 설정하고, 뒤쪽에 0을 채우는 `post` 패딩을 사용합니다.

```python
max_len = 10
pad_sequences(encoded_sentences, maxlen=max_len, padding='post')
```

`padding='post'`는 실제 토큰 뒤에 0을 붙인다는 뜻입니다. 반대로 `padding='pre'`는 앞쪽에 0을 붙입니다.


````python
# 4. 패딩 단계입니다.
# 모든 문장의 길이를 10으로 맞춥니다.
max_len = 10

# 짧은 문장은 뒤쪽에 0을 채워 길이를 맞춥니다.
padded_sentences = pad_sequences(encoded_sentences, maxlen=max_len, padding='post')

# 패딩 결과를 출력합니다.
print("패딩 결과:")
print(padded_sentences)
````

## 1.14 임베딩 설정

임베딩 층을 만들려면 두 가지 값이 중요합니다.

| 값 | 의미 |
|---|---|
| `vocab_size` | 단어 사전의 크기입니다. 패딩 토큰 0까지 고려하기 위해 `+1`을 합니다. |
| `embedding_dim` | 각 토큰을 몇 차원 벡터로 표현할지 정합니다. |

원본 노트북은 임베딩 차원을 `8`로 설정합니다.

실제 LLM은 수백~수천 차원의 임베딩을 사용하지만, 여기서는 구조를 이해하기 위한 작은 예제이므로 8차원을 사용합니다.


````python
# 5. 임베딩 설정입니다.
# tokenizer.word_index에는 실제 단어만 들어 있으므로, 0번 패딩 토큰을 위해 1을 더합니다.
vocab_size = len(tokenizer.word_index) + 1

# 각 토큰을 8차원 벡터로 표현합니다.
embedding_dim = 8

# 설정값을 확인합니다.
print("단어 사전 크기:", vocab_size)
print("임베딩 차원:", embedding_dim)
````

## 1.15 Keras Embedding 모델 생성

Keras의 `Embedding` 층은 정수 ID를 벡터로 바꿉니다.

```python
Embedding(input_dim=vocab_size, output_dim=embedding_dim, input_length=max_len)
```

각 인자의 의미는 다음과 같습니다.

| 인자 | 의미 |
|---|---|
| `input_dim` | 입력 가능한 토큰 ID의 개수, 즉 vocabulary 크기입니다. |
| `output_dim` | 출력 벡터의 차원입니다. |
| `input_length` | 입력 시퀀스 길이입니다. 여기서는 10입니다. |

원본 노트북은 `Sequential` 모델에 Embedding 층 하나를 추가하고, `rmsprop` 옵티마이저와 `mse` 손실함수로 compile합니다. 여기서는 임베딩 출력을 확인하는 목적이므로 별도의 학습은 하지 않습니다.


````python
# 간단한 임베딩 모델을 생성합니다.
model = Sequential()

# Embedding 층을 모델에 추가합니다.
model.add(Embedding(input_dim=vocab_size, output_dim=embedding_dim, input_length=max_len))

# 모델을 컴파일합니다.
# 이 예제에서는 예측을 위한 최소 설정이며, 실제 지도학습 목표를 두고 학습하지는 않습니다.
model.compile('rmsprop', 'mse')

# 모델 구조를 확인합니다.
model.summary()
````

## 1.16 임베딩 결과 확인

패딩된 문장을 임베딩 층에 통과시키면 결과 shape은 다음과 같습니다.

```text
(문장 개수, 문장 길이, 임베딩 차원)
```

현재 예제에서는 문장 3개, 문장 길이 10, 임베딩 차원 8이므로 결과 shape은 보통 `(3, 10, 8)`입니다.

즉 각 문장의 각 토큰이 8차원 벡터로 바뀐 것입니다.


````python
# 패딩된 문장을 임베딩 층에 통과시킵니다.
embeddings = model.predict(padded_sentences)

# 전체 임베딩 결과의 shape을 확인합니다.
print("임베딩 결과 shape:", embeddings.shape)

# 첫 번째 문장의 임베딩 결과를 확인합니다.
print("임베딩 결과 (첫 번째 문장):")
print(embeddings[0])
````

## 1.17 토큰 ID와 임베딩 벡터 연결해서 보기

이제 첫 번째 문장을 기준으로 토큰, ID, 임베딩 벡터를 함께 확인해 보겠습니다.

이 표를 보면 다음 흐름이 눈에 들어옵니다.

```text
원문 일부 -> 토큰 -> 숫자 ID -> 임베딩 벡터
```

패딩 위치는 ID가 0이며, 임베딩 층에서는 0번 토큰도 하나의 벡터를 가집니다. 실제 모델에서는 padding mask를 사용해 패딩 위치가 의미 계산에 영향을 덜 주도록 처리하기도 합니다.


````python
# 첫 번째 문장의 토큰 목록입니다.
first_tokens = tokenized_sentences[0]

# 첫 번째 문장의 패딩된 숫자 ID 목록입니다.
first_ids = padded_sentences[0]

# ID를 다시 단어로 보기 위한 역방향 사전을 만듭니다.
id_to_word = {idx: word for word, idx in tokenizer.word_index.items()}
id_to_word[0] = "<PAD>"

# 토큰 위치별로 토큰, ID, 임베딩 벡터 일부를 출력합니다.
for position, token_id in enumerate(first_ids):
    token = id_to_word.get(int(token_id), "<UNK>")
    vector_preview = embeddings[0][position][:4]
    print(f"위치 {position:02d} | 토큰 {token:<8} | ID {int(token_id):2d} | 벡터 앞 4개 {vector_preview}")
````

## 1.19 RAG와 NLP 처리 과정의 관계

PDF 6.5에서는 RAG의 핵심 기술로 자연어처리 과정, 토크나이저, 임베딩을 다시 설명합니다.

RAG에서는 다음 두 종류의 텍스트가 모두 임베딩됩니다.

- 원본 문서 청크
- 사용자 질문

흐름은 다음과 같습니다.

```text
문서 -> 청크 분할 -> 토큰화/인코딩 -> 임베딩 -> 벡터스토어 저장
질문 -> 토큰화/인코딩 -> 임베딩 -> 벡터스토어에서 유사 문서 검색
검색 문서 + 질문 -> 프롬프트 구성 -> LLM 답변 생성
```

따라서 이 장의 토큰화, 인코딩, 임베딩은 단순한 전처리 개념이 아니라 RAG, 검색, 벡터스토어, LangChain 실습 전체의 기초입니다.
