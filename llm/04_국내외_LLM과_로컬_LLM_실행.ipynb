## 4.4 Ollama란?

Ollama는 로컬 PC에서 LLM을 쉽게 실행할 수 있게 해주는 도구입니다.

PDF 교재는 Ollama의 특징을 다음과 같이 정리합니다.

| 특징 | 설명 |
|---|---|
| 로컬 실행 | 클라우드 서비스에 의존하지 않고 LLM을 로컬에서 실행합니다. |
| 데이터 프라이버시 | 사용자 데이터가 외부 서버로 전송되지 않습니다. |
| 성능 최적화 | 로컬 하드웨어, 특히 GPU를 활용할 수 있습니다. |
| 비용 절감 | 클라우드 API 사용 비용을 줄일 수 있습니다. |
| 모델 커스터마이징 | 필요에 따라 모델을 조정하거나 선택할 수 있습니다. |
| 오프라인 사용 | 인터넷 연결 없이도 이미 다운로드한 모델을 사용할 수 있습니다. |

Ollama는 개인 실습, 소규모 챗봇, 로컬 RAG, 사내 폐쇄망 테스트에 적합합니다.

## 4.6 Ollama 명령어 정리

## 4.8 역할 부여 챗봇

`llm2.ipynb`의 첫 번째 예제는 system prompt를 사용해 모델에게 역할을 부여합니다.

예제 역할:

```text
너는 전문 심리 상담가야. 질문에 대한 답은 3줄 이내로 짧게해줘.
```

이처럼 `system` 메시지로 모델의 행동 규칙을 먼저 지정하면, 모델은 사용자 질문에 답할 때 해당 역할을 따르려고 합니다.

이 방식은 PDF 5.1의 LangChain Agents, Prompt Templates, Memory 개념과 연결됩니다. 모델 자체를 바꾸지 않고도 프롬프트로 역할, 말투, 답변 형식을 어느 정도 제어할 수 있습니다.


````python
# 역할을 부여한 챗봇 함수입니다.
def ask_gemma(question):
    # 챗봇의 기본 역할을 system 메시지로 정의합니다.
    chatbot_role = "너는 전문 심리 상담가야. 질문에 대한 답은 3줄 이내로 짧게해줘."

    # Ollama 모델에 system 역할과 user 질문을 함께 전달합니다.
    response = ollama.chat(
        model='gemma4:e2b',
        messages=[
            {"role": "system", "content": chatbot_role},
            {"role": "user", "content": question},
        ]
    )

    # 모델 응답 텍스트만 반환합니다.
    return response['message']['content']
````

````python
# 역할 부여 챗봇을 호출합니다.
question = "행복하기 위해 어떻게 하면 좋을까?"
response = ask_gemma(question)
print(response)
````

## 4.9 LangChain에서 Ollama 사용

PDF 5.2의 LangChain 사용 예제는 Ollama 모델을 LangChain LLM 래퍼로 감싸서 사용합니다.

원본 코드:

```python
from langchain_community.llms import Ollama
llm = Ollama(model="gemma2")
llm.invoke("안녕, 간단하게 소개해줘.")
```

여기서 중요한 개념은 `invoke()`입니다.

- `invoke()`는 LangChain 객체에 입력을 넣어 실행하는 표준 메서드입니다.
- LLM에 문자열을 넣으면 모델 응답을 반환합니다.
- 이후 RAG 체인에서도 `rag_chain.invoke(query)`처럼 사용합니다.


````python
# LangChain과 Ollama 연동에 필요한 패키지 설치 예시입니다.

# %pip install langchain langchain-community
````

````python
# LangChain의 Ollama LLM 래퍼를 불러옵니다.
from langchain_community.llms import Ollama

# 로컬 Ollama의 gemma2 모델을 LangChain LLM 객체로 연결합니다.
llm = Ollama(model="gemma2")

# invoke()로 프롬프트를 전달하고 응답을 받습니다.
llm.invoke("안녕, 간단하게 소개해줘.")
````

## 4.10 ChatOllama와 메시지 기반 대화

`llm2.ipynb`의 Gradio 챗봇 예제는 `ChatOllama`를 사용합니다.

`Ollama`와 `ChatOllama`의 차이를 간단히 보면 다음과 같습니다.

| 객체 | 입력 형태 | 주 사용처 |
|---|---|---|
| `Ollama` | 문자열 프롬프트 | 단순 LLM 호출, LCEL 체인 |
| `ChatOllama` | `HumanMessage`, `AIMessage` 등 메시지 목록 | 대화형 챗봇, 대화 이력 포함 |

대화형 챗봇에서는 이전 사용자 말과 AI 답변을 함께 넣어야 문맥을 이어갈 수 있습니다.


````python
# ChatOllama와 메시지 객체를 불러옵니다.
from langchain_community.chat_models import ChatOllama
from langchain_core.messages import HumanMessage, AIMessage

# ChatOllama 모델을 초기화합니다.
# temperature가 낮을수록 답변이 안정적이고, 높을수록 다양하고 창의적인 답변이 나올 수 있습니다.
chat_model = ChatOllama(model="gemma4:e2b", temperature=0.7, verbose=False)

# 메시지 목록을 구성합니다.
messages = [
    HumanMessage(content="인공지능에 대해 한 문단으로 설명해줘.")
]

# 메시지를 모델에 전달합니다.
chat_response = chat_model.invoke(messages)

# 응답 내용을 출력합니다.
print(chat_response.content)
````

## 4.11 temperature란?

`llm2.ipynb`의 ChatOllama 예제에는 다음 설정이 등장합니다.

```python
temperature=0.7
```

temperature는 모델 출력의 다양성과 무작위성을 조절하는 값입니다.

| temperature | 경향 |
|---|---|
| 낮음, 예: 0 또는 0.1 | 더 안정적이고 반복적인 답변 |
| 중간, 예: 0.7 | 적당히 자연스럽고 다양한 답변 |
| 높음, 예: 1 이상 | 더 창의적이지만 엉뚱할 가능성 증가 |

주의: `llm2.ipynb`의 한 셀에는 `tempeature=0.1`처럼 오타가 있습니다. 실제 코드에서는 `temperature`로 써야 합니다.


## 4.12 Ollama를 RAG에서 사용하는 이유

이후 RAG 장에서 Ollama는 답변 생성 모델로 사용됩니다.

RAG 흐름에서 Ollama의 위치:

```text
문서 로드 -> 청크 분할 -> 임베딩 -> 벡터스토어 저장
사용자 질문 -> Retriever가 관련 문서 검색
검색 문서 + 질문 -> Ollama LLM이 답변 생성
```

원본 노트북의 RAG 예제에서는 다음처럼 Ollama를 초기화합니다.

```python
llm = Ollama(model="gemma2", base_url="http://localhost:11434")
```

`base_url`은 Ollama 로컬 서버 주소입니다. Ollama가 실행 중이 아니면 연결 오류가 발생합니다.


````python
# RAG에서 사용할 Ollama LLM 초기화 예시입니다.
rag_llm = Ollama(
    model="gemma2",
    base_url="http://localhost:11434"
)
````

## 4.13 Ollama vs vLLM

PDF 5.4의 Ollama vs vLLM 비교를 정리하면 다음과 같습니다.

| 항목 | Ollama | vLLM |
|---|---|---|
| 설치 난이도 | 쉬움, 단일 명령어로 설치 | 중간, Linux/GPU/CUDA 설정 필요 |
| 자원 요구 | CPU 가능, 16GB RAM 이상, GPU 선택 | GPU 필수, 12GB VRAM 이상 |
| 폐쇄형 환경 | 사전 모델 다운로드로 완전 오프라인 가능 | 사전 설치로 오프라인 가능하지만 설정 복잡 |
| 성능 | 소규모 요청에 적합, 속도 보통 | 고속 추론, 다중 요청 처리 우수 |
| 코드 복잡성 | REST API 호출로 간단 | OpenAI API 호환, 약간 더 복잡 |
| 사용 사례 | 개인 프로젝트, 소규모 챗봇 | 상용 서비스, 대규모 요청 처리 |
| 보안 | 로컬 실행, 데이터 외부 전송 없음 | 로컬 실행, 데이터 외부 전송 없음 |

정리하면 다음과 같습니다.

- 학습과 개인 실습: Ollama가 쉽고 빠릅니다.
- 많은 사용자 요청을 처리하는 상용 서비스: vLLM이 더 적합할 수 있습니다.
- GPU 서버와 CUDA 설정에 익숙하지 않다면 Ollama부터 시작하는 것이 좋습니다.

## 4.15 로컬 LLM 사용 시 장점과 한계

### 장점

- 외부 API 비용을 줄일 수 있습니다.
- 민감한 데이터를 외부로 보내지 않을 수 있습니다.
- 인터넷 연결 없이 실습할 수 있습니다.
- 여러 모델을 자유롭게 바꿔 테스트할 수 있습니다.
- RAG, DB 챗봇, 파일 분석 같은 로컬 데이터 실습에 적합합니다.

### 한계

- PC 성능에 따라 속도가 느릴 수 있습니다.
- 큰 모델은 RAM/VRAM 요구량이 큽니다.
- 클라우드 최신 대형 모델보다 성능이 낮을 수 있습니다.
- 모델 설치와 디스크 용량 관리가 필요합니다.
- 멀티모달 지원 여부는 모델마다 다릅니다.

교육 과정에서는 비용과 보안 부담이 적고 반복 실습이 쉬운 로컬 LLM이 매우 유용합니다.


## 4.16 자주 발생하는 오류와 해결 방법

### 1. Ollama 연결 실패

가능한 원인:

- Ollama가 실행 중이 아님
- `localhost:11434` 서버가 열려 있지 않음
- 방화벽 또는 보안 프로그램 문제

해결:

- Ollama 앱을 실행합니다.
- 터미널에서 `ollama list`를 실행해 응답이 오는지 확인합니다.
- 필요하면 `ollama serve`를 실행합니다.

### 2. 모델이 없다는 오류

가능한 원인:

- 코드에 적은 모델명이 로컬에 설치되어 있지 않음
- 모델명 오타

해결:

```powershell
ollama list
ollama pull gemma2
```

### 3. 응답이 너무 느림

가능한 원인:

- 모델이 PC 성능에 비해 큼
- CPU만 사용 중
- 긴 프롬프트 또는 긴 RAG 문맥

해결:

- 더 작은 모델을 사용합니다.
- RAG에서 검색 문서 수 `k`를 줄입니다.
- chunk size를 조정합니다.

### 4. ChatOllama import 오류

해결:

```python
%pip install langchain langchain-community
```

설치 후 커널을 재시작합니다.

### 5. `temperature` 오타

`llm2.ipynb`의 한 코드에는 `tempeature=0.1` 오타가 있습니다. 실제 사용 시 `temperature=0.1`로 수정해야 합니다.


## 4.17 이 장의 핵심 정리

이 장에서 배운 내용을 정리하면 다음과 같습니다.

- LLM은 기업과 모델마다 강점이 다릅니다.
- 해외 LLM은 멀티모달, 긴 컨텍스트, 코딩, 에이전트 기능 등에서 경쟁합니다.
- 국내 LLM은 한국어, B2B, 도메인 특화, 경량화 측면에서 중요합니다.
- Ollama는 로컬에서 LLM을 쉽게 실행하는 도구입니다.
- Ollama의 장점은 프라이버시, 비용 절감, 오프라인 사용, 간편한 실습입니다.
- `ollama pull`, `ollama run`, `ollama list`로 모델을 관리합니다.
- Python `ollama` 패키지로 직접 모델을 호출할 수 있습니다.
- LangChain의 `Ollama`와 `ChatOllama`로 LLM을 체인과 챗봇에 연결할 수 있습니다.
- `invoke()`는 LangChain 객체를 실행하는 기본 메서드입니다.
- Ollama는 개인 실습과 소규모 챗봇에 적합하고, vLLM은 대규모 서비스 추론에 더 적합할 수 있습니다.

다음 장에서는 LangChain 기본 구조를 다룹니다. Prompt, Chain, Memory, LCEL을 이해해야 이후 RAG와 챗봇 실습을 자연스럽게 연결할 수 있습니다.
