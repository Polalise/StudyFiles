```python
# 교재 위치: Appendix B - Gradio 반환값 개수 예시
# 이 셀은 Gradio 함수가 여러 출력을 반환할 때 tuple을 사용하는 예시입니다.

# demo_function 함수는 두 개의 문자열 출력을 반환합니다.
def demo_function(text: str) -> tuple[str, str]:
    # 첫 번째 출력은 원본 입력입니다.
    first_output = text

    # 두 번째 출력은 글자 수 정보입니다.
    second_output = f"글자 수: {len(text)}"

    # 두 개의 출력값을 tuple로 반환합니다.
    return first_output, second_output

# 예시 함수를 실행해 반환값 형태를 확인합니다.
print(demo_function("테스트"))
```

## B.8 DB 챗봇 오류

| 오류 상황 | 원인 | 해결 방법 |
|---|---|---|
| DB 연결 실패 | URL, 계정, 비밀번호, 포트 오류 | DB_URL과 접속 권한 확인 |
| 테이블을 찾을 수 없음 | 스키마 또는 DB명 불일치 | 현재 접속 DB와 테이블 목록 확인 |
| SQL 문법 오류 | LLM이 잘못된 SQL 생성 | 스키마 정보를 프롬프트에 정확히 제공 |
| 위험한 SQL 생성 | LLM이 INSERT/DELETE 생성 | SELECT만 허용하는 검증 함수 추가 |
| 결과가 너무 많음 | LIMIT 없음 | 프롬프트에 LIMIT 사용 지시 |
| 한글 깨짐 | DB 인코딩 문제 | DB 문자셋과 연결 옵션 확인 |

DB 챗봇에서는 LLM이 생성한 SQL을 바로 실행하지 말고, 반드시 검증 단계를 거쳐야 합니다.


```python
# 교재 위치: Appendix B - 안전한 SQL 검사 함수
# 이 셀은 DB 챗봇에서 사용할 수 있는 기본 SQL 안전성 검사 예시입니다.

# is_safe_sql 함수는 SQL이 SELECT 문인지 확인합니다.
def is_safe_sql(sql: str) -> bool:
    # SQL이 비어 있으면 안전하지 않다고 판단합니다.
    if not sql:
        # 비어 있는 SQL은 실행 대상이 아닙니다.
        return False

    # SQL을 소문자로 바꿔 키워드 검사를 쉽게 합니다.
    lowered_sql = sql.lower()

    # 차단할 위험 키워드 목록입니다.
    blocked_keywords = ["insert", "update", "delete", "drop", "alter", "truncate", "create"]

    # 위험 키워드 포함 여부를 확인합니다.
    has_blocked_keyword = any(keyword in lowered_sql for keyword in blocked_keywords)

    # SELECT로 시작하고 위험 키워드가 없으면 안전하다고 판단합니다.
    return lowered_sql.strip().startswith("select") and not has_blocked_keyword


# 안전한 SQL 예시를 검사합니다.
print(is_safe_sql("SELECT * FROM customers LIMIT 10"))

# 위험한 SQL 예시를 검사합니다.
print(is_safe_sql("DROP TABLE customers"))
```

음성 실습은 외부 환경 영향을 많이 받습니다. 먼저 짧고 깨끗한 음성 파일로 테스트하는 것이 좋습니다.

## B.15 디버깅 체크리스트

오류가 발생하면 다음 순서로 확인합니다.

3. 해당 셀에서 사용하는 패키지가 설치되어 있는지 확인합니다.
4. 현재 노트북 커널이 올바른 Python 환경인지 확인합니다.
5. 입력 파일 경로가 맞는지 확인합니다.
6. API 키, DB URL, 모델명 같은 설정값이 맞는지 확인합니다.
7. LLM 호출 전에 프롬프트 입력값을 먼저 출력해 확인합니다.
8. RAG에서는 검색 결과 문서를 먼저 출력해 확인합니다.
9. DB 챗봇에서는 생성된 SQL을 실행 전에 출력하고 검증합니다.
10. Gradio에서는 함수 반환값 개수와 output 컴포넌트 개수를 확인합니다.
11. 문제가 계속되면 작은 예제 코드로 분리해 다시 실행합니다.

복잡한 LLM 앱의 오류는 한 번에 해결하려고 하기보다, `입력 -> 중간 결과 -> 출력` 순서로 끊어서 확인하는 것이 가장 안정적입니다.


```python
# 교재 위치: Appendix B - 간단한 입력값 점검 함수
# 이 셀은 LLM 함수 실행 전에 입력값이 비어 있는지 확인하는 공통 유틸리티 예시입니다.

# validate_required_text 함수는 필수 텍스트 입력값이 비어 있는지 검사합니다.
def validate_required_text(value: str, field_name: str) -> str:
    # 값이 None이거나 공백만 있으면 오류 메시지를 반환합니다.
    if value is None or not str(value).strip():
        # 어떤 필드가 비어 있는지 알려주는 메시지입니다.
        return f"{field_name} 입력값이 비어 있습니다."

    # 입력값이 정상이라면 빈 문자열을 반환합니다.
    return ""


# 질문 입력값을 검사하는 예시입니다.
validation_message = validate_required_text("", "질문")

# 검사 결과를 출력합니다.
print(validation_message)
```
